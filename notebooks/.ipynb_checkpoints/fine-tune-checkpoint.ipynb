```python
pipe.unet.train()
pipe.text_encoder.train()
pipe.text_encoder_2.train()
pipe.vae.train()
```

```python
dataloader = caption_dataset()
train_dataloader,_ = dataloader.get_dataloader(partition="train", batch_size=32)
```

```python
device = "cuda"

pipe.to(device)
vae = pipe.vae
unet = pipe.unet
tokenizer = pipe.tokenizer
tokenizer_2 = pipe.tokenizer_2
text_encoder = pipe.text_encoder
text_encoder_2 = pipe.text_encoder_2
noise_scheduler = pipe.scheduler
```

```python
optimizer = torch.optim.AdamW(pipe.unet.parameters(), lr=5e-6)
loss_fn = nn.MSELoss()
```

```python
num_epochs = 5
```

```python
proj_text_to_add = nn.Linear(1280,  (unet.add_embedding.linear_1.in_features)).to(device)

optimizer = torch.optim.AdamW(
    list(unet.parameters()) + list(proj_text_to_add.parameters()),
    lr=1e-5
)
loss_fn = nn.MSELoss()

for epoch in range(num_epochs):
    for imgs, texts in train_dataloader:
        imgs = imgs.to(device)

        # encode images -> latents
        latents = vae.encode(imgs).latent_dist.sample() * 0.18215
        
        t = torch.randint(0, noise_scheduler.num_train_timesteps, (latents.shape[0],), device=device).long()
        print(t.shape)

        # sample noise and timesteps
        noise = torch.randn_like(latents)
        
        noisy_latents = noise_scheduler.add_noise(latents, noise, t)
        
        # encode text (two encoders)
        inputs_1 = tokenizer(list(texts), padding="max_length", truncation=True, return_tensors="pt").to(device)
        inputs_2 = tokenizer_2(list(texts), padding = "max_length", truncation = True, return_tensors = "pt").to(device)
        text_enc_out_1 = text_encoder(**inputs_1)
        text_enc_out_2 = text_encoder_2(**inputs_2)
        
        prompt_embeds_1 = text_enc_out_1.last_hidden_state          # (B, seq_len, H)
        prompt_embeds_2 = text_enc_out_2.last_hidden_state          # (B, seq_len, H)
        print(prompt_embeds_1.shape)
        print(prompt_embeds_2.shape)
        
        #projected_embeds_2 = proj_text_to_add(prompt_embeds_2)
        projected_embeds_2 = prompt_embeds_2.mean(dim=1)  # (B, H_text)
        proj_text_to_add = nn.Linear(1280, (unet.add_embedding.linear_1.in_features- 256)).to(device)
        projected_embeds_2 = proj_text_to_add(projected_embeds_2)
        print(projected_embeds_2.shape)
        # Prepare added_cond_kwargs
        time_ids = t.unsqueeze(-1)  # (B, 1)

        # project using UNet's linear
        time_embeds = unet.add_time_proj(time_ids.flatten())  # (B, H_time))

        print(time_embeds.shape)
        added_cond_kwargs = {
            "text_embeds": projected_embeds_2,
            "time_ids": time_ids
        }

        # Forward pass through UNet
        noise_pred = unet(
            noisy_latents,
            t,
            encoder_hidden_states=prompt_embeds_1,
            added_cond_kwargs=added_cond_kwargs
        ).sample

        # Compute loss (MSE between predicted noise and true noise)
        loss = loss_fn(noise_pred, noise)

        # Backprop
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {loss.item():.4f}")
        
```
